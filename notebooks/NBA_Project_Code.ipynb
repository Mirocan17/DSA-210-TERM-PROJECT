```python
import pandas as pd
import time

years = range(1994, 2025)

all_drafts = []

print("Data processing..")

for year in years:
    url = f"https://www.basketball-reference.com/draft/NBA_{year}.html"

    try:
        dfs = pd.read_html(url, header=[0, 1])
        df = dfs[0]

        if 'Advanced' in df.columns.get_level_values(0):
            df = df.drop('Advanced', axis=1, level=0)

        df = df[df[('Round 1', 'Player')] != 'Player']

        df['Year'] = year

        all_drafts.append(df)
        print(f"{year} is done")

        time.sleep(3)

    except Exception as e:
        print(f"{year} problem: {e}")

final_df = pd.concat(all_drafts, ignore_index=True)

final_df.to_csv('nba_drafts_1994_2024_cleaned.csv', index=False)
print("\nJobs done.")

from google.colab import files
files.download('nba_drafts_1994_2024_cleaned.csv')
```

```python
import pandas as pd
import time
from google.colab import files
years = range(1994, 2025)

team_conference_map = {
    # === DOU (EAST) ===
    'Boston Celtics': 'East',
    'Brooklyn Nets': 'East',
    'New Jersey Nets': 'East', # Eski isim
    'New York Knicks': 'East',
    'Philadelphia 76ers': 'East',
    'Toronto Raptors': 'East',
    'Chicago Bulls': 'East',
    'Cleveland Cavaliers': 'East',
    'Detroit Pistons': 'East',
    'Indiana Pacers': 'East',
    'Milwaukee Bucks': 'East',
    'Atlanta Hawks': 'East',
    'Charlotte Hornets': 'East',
    'Charlotte Bobcats': 'East', # Eski isim
    'Miami Heat': 'East',
    'Orlando Magic': 'East',
    'Washington Wizards': 'East',
    'Washington Bullets': 'East', # 1997 ncesi isim

    # === BATI (WEST) ===
    'Denver Nuggets': 'West',
    'Minnesota Timberwolves': 'West',
    'Oklahoma City Thunder': 'West',
    'Seattle SuperSonics': 'West', # EFSANE (OKC'nin eski hali)
    'Portland Trail Blazers': 'West',
    'Utah Jazz': 'West',
    'Golden State Warriors': 'West',
    'Los Angeles Clippers': 'West',
    'Los Angeles Lakers': 'West',
    'Phoenix Suns': 'West',
    'Sacramento Kings': 'West',
    'Dallas Mavericks': 'West',
    'Houston Rockets': 'West',
    'Memphis Grizzlies': 'West',
    'Vancouver Grizzlies': 'West', # 2001 ncesi Kanada'dayken
    'New Orleans Pelicans': 'West',
    'New Orleans Hornets': 'West', # Eski isim
    'Oklahoma City Hornets': 'West', # Katrina Kasrgas dnemi geici isim
    'San Antonio Spurs': 'West'
}

all_data = []
print("With the old names...")

for year in years:
    url = f"https://www.basketball-reference.com/leagues/NBA_{year}_standings.html"

    try:
        dfs = pd.read_html(url)
        found_table = False

        for df in dfs:
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = df.columns.map(' '.join).str.strip()

            df = df.rename(columns={df.columns[0]: 'Team'})

            if 'W' in df.columns and 'L' in df.columns and len(df) > 10:

                df = df[~df['Team'].str.contains("Division", na=False)]
                df = df[~df['Team'].str.contains("Conference", na=False)]

                df['Team'] = df['Team'].str.replace(r'\*|\(\d+\)', '', regex=True).str.strip()

                df['Year'] = year

                df['Conference'] = df['Team'].map(team_conference_map)

                if df['Conference'].notna().sum() > 10:

                    cols_to_keep = ['Year', 'Conference', 'Team', 'W', 'L', 'W/L%']
                    valid_cols = [c for c in cols_to_keep if c in df.columns]
                    df = df[valid_cols]

                    all_data.append(df)
                    found_table = True

        if found_table:
            print(f"{year} done.")
        else:
            print(f"{year} could not done")

        time.sleep(3)

    except Exception as e:
        print(f"{year} error: {e}")

if all_data:
    final_df = pd.concat(all_data, ignore_index=True)

    final_df = final_df.drop_duplicates(subset=['Year', 'Team'])

    final_df['W'] = pd.to_numeric(final_df['W'], errors='coerce')
    final_df['L'] = pd.to_numeric(final_df['L'], errors='coerce')

    final_df = final_df.sort_values(by=['Year', 'Conference', 'W'], ascending=[True, True, False])

    filename = 'nba_standings_1994_2025_robust.csv'
    final_df.to_csv(filename, index=False)

    print(f"\nProcessed: {filename}")
    files.download(filename)
else:
    print("Something went wrong.")
```

```python
import pandas as pd
import numpy as np
from google.colab import files

input_filename = 'nba_standings_1994_2025_robust.csv'

ODDS_OLD = [250, 199, 156, 119, 88, 63, 43, 28, 17, 11, 8, 7, 6, 5]

ODDS_NEW = [140, 140, 140, 125, 105, 90, 75, 60, 45, 30, 20, 15, 10, 5]

try:
    df = pd.read_csv(input_filename)
    print(f"{input_filename} is read successfully.")

    all_lottery_data = []

    years = sorted(df['Year'].unique())

    print("Calculating...")

    for year in years:
        season_df = df[df['Year'] == year].copy()

        season_df = season_df.sort_values(by=['W', 'L'], ascending=[True, False]).reset_index(drop=True)

        lottery_teams = season_df.head(14).copy()

        current_odds = ODDS_NEW if year >= 2019 else ODDS_OLD

        lottery_teams['Win_Prob_%'] = 0.0

        i = 0
        while i < len(lottery_teams):
            wins = lottery_teams.loc[i, 'W']

            count = 0
            for j in range(i, len(lottery_teams)):
                if lottery_teams.loc[j, 'W'] == wins:
                    count += 1
                else:
                    break

            indices = range(i, i + count)

            valid_indices = [idx for idx in indices if idx < len(current_odds)]

            if valid_indices:
                total_odds = sum([current_odds[idx] for idx in valid_indices])
                avg_prob = total_odds / len(valid_indices)

                lottery_teams.loc[indices, 'Win_Prob_%'] = avg_prob / 10.0

            i += count

        all_lottery_data.append(lottery_teams)
        print(f"{year} is done.")

    final_lottery_df = pd.concat(all_lottery_data, ignore_index=True)

    final_lottery_df = final_lottery_df[['Year', 'Team', 'W', 'L', 'Win_Prob_%']]

    output_filename = 'nba_calculated_lottery_odds.csv'
    final_lottery_df.to_csv(output_filename, index=False)

    print(f"\n Data is ready: {output_filename}")
    files.download(output_filename)

except FileNotFoundError:
    print(f"HATA: '{input_filename}' cannot be found!")
```
